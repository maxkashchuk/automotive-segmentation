```python
import cv2 as cv
import matplotlib.pyplot as plt

img_path = r"..\..\visualization_resources\line_example.jpg"
img = cv.imread(img_path)
assert img is not None, "Image file could not be read"

coordinates_file = r"..\..\visualization_resources\line_example.lines.txt"

lines = []

with open(coordinates_file, 'r') as file:
    lines = [list(map(int, map(float, line.split()))) for line in file]

r, g, b = 255, 255, 255

thickness = 2

i = 0

line_id = 0

for line in lines:
    i = 0

    while i < len(line) - 2:
        x1, y1 = line[i], line[i + 1]
        x2, y2 = line[i + 2], line [i + 3]

        cv.line(img, (int(x1), int(y1)), (int(x2), int(y2)), (r, g, b), thickness)

        i += 2

    line_id += 1

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))

plt.subplot()
plt.title('Lines annotation visualization')
plt.imshow(img)

plt.tight_layout()
plt.show()
```
